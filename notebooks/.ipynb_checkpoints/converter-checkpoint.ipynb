```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image1 = cv2.imread("../app/assets/Imagens_teste/2817540617.jpg")
```

```python
%matplotlib inline 
# or in colab you can use : from google.colab.patches import cv2_imshow
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
plt.imshow(image1)
plt.show()
```

## RGB TO YIQ

```python
plt.imshow(image1);
```

## Converting from rgb to YIQ

```python
(b, g, r) = cv2.split(image1)
```

```python
import skimage
```

```python
Y = 0.299 * r + 0.587 * g + 0.114 * b
I = 0.596 * r - 0.274 * g - 0.322 * b
Q = 0.211 * r - 0.523 * g + 0.312 * b
YIQ = cv2.merge([Y, I, Q]) / 255.0
im1 = skimage.color.rgb2yiq(image1)
print(im1[0][:10])
print("\n")
print(YIQ[0][:10])
plt.imshow(YIQ)
plt.show()
plt.imshow(im1)
```

## Converting YIQ to RGB

```python
(y, i, q) = cv2.split(im1)
```

```python
r = y + 0.956 * i + 0.621 * q
g = y - 0.272 * i - 0.647 * q
b = y - 1.106 * i + 1.703 * q
```

```python
rgb_im1 = cv2.merge([b, g, r])
```

```python
plt.imshow(rgb_im1)
plt.show()
plt.imshow(image1)
plt.show()
plt.imshow(skimage.color.yiq2rgb(im1))
```
